```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
def plot_graph(adjacency_list, colored_nodes):
    G = nx.Graph(adjacency_list)
    pos = nx.spring_layout(G)  # You can use different layout algorithms

    # Specify colors for nodes
    node_colors = ['red' if node in colored_nodes else 'skyblue' for node in G.nodes]

    # Draw the graph
    nx.draw(G, pos, with_labels=True, font_weight='bold', node_size=700, node_color=node_colors, font_color='black', font_size=8)

    plt.show()
    
    
def getGraph(edges):
    Graph = {}
    
    for nodeA,nodeB in edges: 
        
        if nodeA not in Graph:
            Graph[nodeA] = []
        if nodeB not in Graph:
            Graph[nodeB] = []
            
        Graph[nodeA].append(nodeB)
        Graph[nodeB].append(nodeA)
    
    
    return Graph
```

```python
edges = [['w','x'],['x','y'],['z','y'],['z','v'],['w','v']]


adjacency_list = getGraph(edges=edges)

plt.figure(dpi=150)
plot_graph(adjacency_list,['w','y'])
```

```python
# For this the Breadth First Traversal algorithm is more convenient as it explores 
# more evenly all the different directions.

def countShortPartBFT(graph,start,destination):
    
    
    queue = [[start,0]]
    visited = set(start)
    
    while queue:
        
        current,count = queue.pop(0)
        
        if current == destination:
            return count
        
        for neighbor in graph[current]:
            if neighbor not in visited:
                queue.append([neighbor,count+1])
                visited.add(neighbor)
    
    return -1


print('The shortest path is : ', countShortPartBFT(adjacency_list,'y','w'))
print('The shortest path is : ', countShortPartBFT(adjacency_list,'y','a'))
```
